```python
df.drop(['2401','Borderlands'],axis=1,inplace=True)
```

```python
df.rename(columns={'Positive':'Target','im getting on borderlands and i will murder you all ,':'Tweets'},inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()
df['Target'] = la.fit_transform(df['Target'])
```

```python
df.head()
```

```python
df['Tweets'] = df['Tweets'].astype(str)
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
```

```python
def remove_stopwords_punctuations(text):

  # Tokenize the text into words
  tokens = text.split()

  # Remove punctuations
  tokens = [token for token in tokens if token not in string.punctuation]

  # Remove stopwords
  stop_words = set(stopwords.words('english'))
  tokens = [token for token in tokens if token.lower() not in stop_words]

  # Join the tokens back into a string
  text_without_stopwords_punctuations = " ".join(tokens)
  return text_without_stopwords_punctuations
```

```python
df['Tweets'] = df['Tweets'].apply(remove_stopwords_punctuations)
```

```python
from sklearn.feature_extraction.text import CountVectorizer # Corrected the import statement to use the right module name.
cv = CountVectorizer()
x = cv.fit_transform(df['Tweets']).toarray()
```

```python
x
```

```python
x = df["Tweets"]
y = df["Target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)*100
```
